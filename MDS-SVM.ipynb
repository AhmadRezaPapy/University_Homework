```python
import pandas as pd

embedding = pd.read_csv(r'/content/sample_data/mds15data.csv')
df2 = pd.read_csv('/content/sample_data/labels data.csv')
```

```python
df2.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
X_train, X_test, y_train, y_test = train_test_split(embedding, df2['class'], test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
svm_model = SVC(kernel='linear')  # You can also use 'rbf', 'poly', etc. for the kernel
svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test,y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
# auc_roc = roc_auc_score(y_test,y_pred, multi_class='ovr')
conf_mat = confusion_matrix(y_test, y_pred)

# چاپ معیارها
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
# print(f"AUC-ROC: {auc_roc}")
print(f"Confusion Matrix:\n{conf_mat}")
```
