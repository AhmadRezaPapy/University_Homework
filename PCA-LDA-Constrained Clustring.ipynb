```python
import pandas as pd
import numpy as np

# فرض کنید df یک DataFrame با 10000 سطر و 7 ستون ویژگی و یک ستون برچسب وجود دارد
np.random.seed(42)  # برای تولید نتایج قابل پیش‌بینی
df = pd.DataFrame(
    np.random.randint(0, 10, size=(10000, 8)),  # 8 ستون: 7 ویژگی + 1 برچسب
    columns=['feature'+str(i) for i in range(1, 8)] + ['label']
)
df['label'] = np.random.randint(0, 7, size=10000)  # برچسب‌ها را تصادفی تولید می‌کنیم

# تعداد نمونه‌ها
num_samples = 1000

# تعداد کلاس‌ها
num_classes = 7

# تعداد نمونه‌هایی که باید در هر گروه انتخاب شوند
samples_per_group = num_samples // num_classes

# تعداد گروه‌ها
groups = num_samples // samples_per_group

# ایجاد یک DataFrame خالی برای ذخیره نمونه‌های انتخاب شده
selected_df = pd.DataFrame()

# انتخاب نمونه‌ها
for i in range(groups):

    group = df.sample(n=samples_per_group, random_state=i)
    selected_df = pd.concat([selected_df, group])


# چاپ نمونه‌های انتخاب شده

print("\nنمونه‌های انتخاب شده:")
display(selected_df)

```

الگوریتم **LDA**

```python
import pandas as pd
import numpy as np

def calculate_class_means(data, labels):
    unique_labels = np.unique(labels)
    class_means = {}
    for label in unique_labels:
        class_data = data[labels == label]
        class_mean = np.mean(class_data, axis=0)
        class_means[label] = class_mean
    return class_means
def calculate_S_B(data, labels, class_means):
    n_classes = len(class_means)
    S_B = np.zeros((data.shape[1], data.shape[1]))
    for i in range(n_classes):
        mu_i = class_means[i]
        mu = np.mean(data, axis=0)
        S_B += len(data[labels == i]) * ((mu_i - mu) @ (mu_i - mu).T)
    return S_B
def calculate_S_W(data, labels):
    n_classes = len(np.unique(labels))
    S_W = np.zeros((data.shape[1], data.shape[1]))
    for i in range(n_classes):
        class_data = data[labels == i]
        S_W += np.cov(class_data, rowvar=False)
    return S_W
def calculate_S_W(data, labels):
    n_classes = len(np.unique(labels))
    S_W = np.zeros((data.shape[1], data.shape[1]))
    for i in range(n_classes):
        class_data = data[labels == i]
        S_W += np.cov(class_data, rowvar=False)
    return S_W
def calculate_B(S_W, S_B):
    # اضافه کردن یک مقدار ثابت به S_W
    S_W = S_W + np.eye(S_W.shape[0])
    return np.linalg.inv(S_W) @ S_B

def calculate_eigvals_eigvecs(B):
    eigvals, eigvecs = np.linalg.eig(B)

    return eigvals, eigvecs

def sort_eigvecs_eigvals(eigvals, eigvecs):
    sort_indices = np.argsort(eigvals)[::-1]
    sorted_eigvals = eigvals[sort_indices]
    sorted_eigvecs = eigvecs[:, sort_indices]

    return sorted_eigvals, sorted_eigvecs

def reduce_dimensions(data, sorted_eigvecs, num_dimensions):
    # Ensure num_dimensions does not exceed the number of columns in sorted_eigvecs
    if num_dimensions > sorted_eigvecs.shape[1]:
        raise ValueError(f"num_dimensions {num_dimensions} is greater than the number of eigenvectors {sorted_eigvecs.shape[1]}")
    # Perform the dot product only if the dimensions match
    reduced_data = data.dot(sorted_eigvecs[:, :num_dimensions])
    return reduced_data

num_dimensions = 3


# استخراج داده‌ها و برچسب‌ها
data = selected_df.iloc[:, :-1].values
labels = selected_df.iloc[:, -1].values

# محاسبه میانگین کلاس‌ها
class_means = calculate_class_means(data, labels)

# محاسبه S_B
S_B = calculate_S_B(data, labels, class_means)

# محاسبه S_W
S_W = calculate_S_W(data, labels)

# محاسبه B
B = calculate_B(S_W, S_B)

# محاسبه مقادیر و بردارهای ویژه
eigvals, eigvecs = calculate_eigvals_eigvecs(B)

# مرتب‌سازی مقادیر و بردارهای ویژه
sorted_eigvals, sorted_eigvecs = sort_eigvecs_eigvals(eigvals, eigvecs)

# کاهش ابعاد
reduced_dimensions = reduce_dimensions(data, sorted_eigvecs, num_dimensions)


display(reduced_dimensions)
```

خوشه بندی **مقید**

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN



# داده‌های نمونه
data =reduced_dimensions

X = data[:, :-1]  # فرض بر این است که ستون آخر لیبل‌ها هستند
true_classes = data[:, -1]

n_samples = X.shape[0]
n_classes = len(np.unique(true_classes))
total_variance = np.var(X)
must_link = np.zeros((n_samples, n_samples))
mustnot_link = np.zeros((n_samples, n_samples))
k = 7
cluster_radii = np.zeros(k)
kmeans = KMeans(n_clusters=k, init='k-means++')
kmeans.fit(X, sample_weight=must_link.sum(axis=0))
cluster_centers = kmeans.cluster_centers_
# ایجاد نمونه‌ای از کلاس و آموزش آن با واریانس کل داده‌ها


for i in range(k):
    cluster_points = X[kmeans.labels_ == i]
    cluster_radii[i] = np.mean(np.sqrt(np.sum((cluster_points - cluster_centers[i])**2, axis=1)))

for i in range(n_samples):
    for j in range(i+1, n_samples):
        for i in range(k):
            if cluster_radii[i] <= total_variance:
               must_link[i, j] = must_link[j, i] = 1



for i in range(n_samples):
    for j in range(i+1, n_samples):
         for i in range(k):
             if cluster_radii[i] >= total_variance:
                mustnot_link[i, j] = mustnot_link[j, i] = 1



# استخراج خوشه‌ها با DBSCAN و قید mustn't link
dbscan = DBSCAN(eps=1.5, min_samples=2)
dbscan.fit(X, sample_weight=-mustnot_link.sum(axis=0))

# انتخاب خوشه‌های مشترک بین دو روش خوشه‌بندی
final_clusters = []
for i in range(k):
    kmeans_indices = np.where(kmeans.labels_ == i)[0]
    dbscan_indices = np.where(dbscan.labels_ == i)[0]
    common_indices = list(set(kmeans_indices).intersection(dbscan_indices))
    final_clusters.append(common_indices)

# پیشبینی توسط مدل K-means
predicted_clusters = np.zeros(X.shape[0], dtype=int)
for i, cluster in enumerate(final_clusters):
    predicted_clusters[cluster] = i

# محاسبه ماتریس درهم ریختگی و نمایش آن
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
# تبدیل برچسب‌های پیوسته به دودویی
le = LabelEncoder()
true_classes_binary = le.fit_transform(true_classes)


# ایجاد DataFrame جدید برای نمایش


# نمایش جدول با استفاده از pandas
display(true_classes_binary)
display(predicted_clusters)
# حالا می‌توانید ماتریس درهم ریختگی را محاسبه کنید
cm = confusion_matrix(true_classes_binary, predicted_clusters)
print(cm)



accuracy = np.mean(true_classes == predicted_clusters)
print('Accuracy of the clustering algorithm:', accuracy)
```
