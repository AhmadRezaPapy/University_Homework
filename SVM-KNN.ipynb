```python
import pandas as pd
file_path = "/content/sample_data/data.csv"
df = pd.read_csv(file_path)
df
```

```python
file_path_1 = "/content/sample_data/labels data.csv"
df2 = pd.read_csv(file_path_1)
df2
```

```python
from sklearn.model_selection import train_test_split


df_cleaned = df.iloc[:, 2:]

# تقسیم‌بندی داده‌ها به آموزش و تست
X_train, X_test, y_train, y_test = train_test_split(df_cleaned, df2['class'], test_size=0.3, random_state=42)

from sklearn.svm import SVC

model = SVC(kernel='linear')  # می‌توانید نوع هسته را تغییر دهید
model.fit(X_train, y_train)
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report

# لیست مقادیر k
ks = [1, 3, 5, 7, 13, 15]

for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    # پیش‌بینی بر روی مجموعه تست
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"k={k}, Accuracy: {accuracy:.2f}")

    # محاسبه Precision و Recall
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')

    print(f"k={k}, Precision: {precision:.2f}, Recall: {recall:.2f}")

    # گزارش جامع ارزیابی
    report = classification_report(y_test, y_pred, target_names=df2['class'].unique())
    print("\nClassification Report:\n", report)
```
